```python
#Consultando linhas com valores faltantes(nulos)
df.isnull().sum()
```

```python
#substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(30)
```

```python
df.isnull().sum()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"],inplace=True)
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all",inplace=True)
```

# **Criando colunas novas**

```python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Criando a coluna de Receita/Vendas
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Rertornando a menor receita
df["Receita"].min()
```

```python
#Retorna as lojas com maiores receitas "nlargest"
df.nlargest(3, "Receita")
```

```python
#Retorna as lojas com menores  receitas "nsmallest"
df.nsmallest(3, "Receita")
```

```python
#agrupamento somandos as receitas por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordernando o conjunto de dados por receita
df.sort_values("Receita", ascending=False).head(30)
```

# **Trabalhando com Datas - 14-10-2022**

```python
#Transformando a coluna da data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo e criando colunas com mês e dia
df["mes_venda"],df["dia_venda"] = df["Data"].dt.month, df['Data'].dt.day
```

```python
df.sample(5)
```

```python
#Retornando data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna trimestre
df["trimestre_vendas"] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
vendas_marco_19.sample(10)
```
