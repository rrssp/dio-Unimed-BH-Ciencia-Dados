# **Analises exploratóris**

```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#leitura dos arquivos e criando nosso dataframe
df = pd.read_excel("/content/drive/MyDrive/Unimed-BH Ciencia de Dados/Datasets/AdventureWorks.xlsx")
```

```python
#Visualizar de 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#verificando tipos de dados
df.dtypes
```

```python
#Qual a receita total?
df['Valor Venda'].sum()
```

```python
#Qual o custo total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criando a coluna de custo
```

```python
df.head(1)
```

```python
#Qual custo total?
round(df["custo"].sum(),2)
```

```python
#Agora que temos a receita e custo e o total, podemos acgar o lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
round(df["lucro"].sum(),2)
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

**Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_Envio em numerica**

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#Media do tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

**Missing Values**

```python
df.isnull().sum()
```

E, se a gente quiser saber o lucro por ano e Por Marca

```python
#Vamos agrupar por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Grafico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="total Produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto")
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mes")
plt.xlabel("Mes")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro");
plt.xticks(rotation="horizontal")
```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro");
plt.xticks(rotation="horizontal")
```
